#### Global infrastructure

19 regions
57 availability zones

An availability zone is a data center
- can have serveral data centers a couple of miles from each other

A data center has:
- a lot of computers, load balancers, firewalls, switches, storage

A region is a geographical area that consist of 2 or more availability zones

Regions can be GovCloud for goverment only

A region has a location, launch date and EC2 availability zones amount

https://aws.amazon.com/about-aws/global-infrastructure/regions_az/


Edge location are endpoints for aws which are used for caching content. Typically this consists of cloudfront, amazon's content delivery network (CDN)

There are many more edge locations than regions. Currently over 150 edge locations
